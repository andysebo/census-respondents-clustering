## 01 Data cleaning and preprocessing

This notebook serves as a data cleaning pipeline for the clustering project.




```python
import pandas as pd

df = pd.read_csv('/Users/andrejsebo/Projects/clustering-project/data/raw/data.txt', sep = '\t')

df.head
```

```python
df_new = df.drop(columns= ['reg', 'vb', 'hinc', 'hmem', 'hint', 'fixline', 'hcar', 'sex', 'agecat', 'marits', 'red', 'empst', 'hhead', 'pinc', 'sim', 'seg_ls'])

df_new.head()
```

Since the dataset contains many columns that are not relevant for our clustering analysis, we will drop them to simplify our dataset and focus on the most important features.

We move on to setting id as the index, now the old dataframe remains floating in the memory, the python garbage collector will eventually clean it up as it sees no vars are pointing to it.


```python
df_new = df_new.set_index('id')
df_new.head()
```

Let's do some data validation now. The plan is to check whether we only go 0's or 1's for all questions and all observations. Then we check for missing values and for obs that only contain either 0's or 1's.

```python
df_new.isin([0, 1]).all().all()

row_sums = df_new.sum(axis=1)
print(row_sums)

#row_sums.value_counts()
```

```python
row_sums = df_new.sum(axis=1)
print(row_sums)

#row_sums.value_counts()
```

```python
id_to_drop = row_sums[(row_sums < 3) | (row_sums > 17)].index
print(id_to_drop)
df_cleaned = df_new.drop(index=id_to_drop)
print(df_new.shape)
print(df_cleaned.shape)
```

Okay so this would be the end of data loading, formatting and cleaning. The only thing left to do is to save and freeze this final dataset for the next steps in the project. We will save it as a csv file in the data/processed folder. We will also save it as a pickle file, which is a binary format that preserves the data types and structure of the dataframe, making it faster to load and use in future steps of the project.

```python
df_final = df_cleaned.to_csv('/Users/andrejsebo/Projects/clustering-project/data/processed/data_cleaned.csv', index=True)

df_final_pickle = df_cleaned.to_pickle('/Users/andrejsebo/Projects/clustering-project/data/processed/data_cleaned.pkl')     
```
